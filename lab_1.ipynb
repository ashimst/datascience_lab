### Step-5: Removing unnecessary variables(columns)

```python
df.drop('Not_Useful_Column',axis=1,inplace=True)
df.drop('Unnamed: 8',axis=1,inplace=True)
df.drop('Unnamed: 9',axis=1,inplace=True)
```

```python
df
```

### Step-6: Cleaning the LastName column
Clean the LastName by stripping the string. Use str.strip("123./").

```python
df['Last_Name']=df['Last_Name'].str.strip("123./")
df
```

### Step-7: Maintain Data Uniformity

```python
# In the column Paying Customer, replace all 'Yes' to 'Y' and all 'No' to 'N' making it uniform.
df['Paying Customer'] = df['Paying Customer'].replace({'Yes': 'Y', 'No': 'N'})
df
```

```python
df['Phone_Number'].info
```

```python
# Formatting the phone number format: (123-456-7890).
df['Phone_Number'] = df['Phone_Number'].str.replace(r'[^0-9]', '', regex=True)  # Remove non-digit characters
df['Phone_Number'] = df['Phone_Number'].str.slice(0, 3) + '-' + df['Phone_Number'].str.slice(3, 6) + '-' + df['Phone_Number'].str.slice(6, 10)
```

### Step-8: Dealing with missing values
1. Replace all 'N/a' with 'NaN' to entire dataset.
2. Fill the 'NaN' data with "".

```python
df.replace('N/a',np.nan,inplace=True)
df
```

```python
df1=df.replace(np.nan,'')
df1
```

### Step-9: Drop the rows that dont want to be contacted

```python
df1['Do_Not_Contact'] = df1['Do_Not_Contact'].replace({'Yes': 'Y', 'No': 'N'})
```

```python
condition= df1['Do_Not_Contact']=='N'
df2= df1.loc[condition]
df2
```
